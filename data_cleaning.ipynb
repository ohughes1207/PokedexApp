```python
import pandas as pd
```

```python
pokemon_df = pd.read_csv('pokemonscraper\pokemon.csv', sep=',')
```

1010 unique values for Pokedex number is expected

```python
pokemon_df['Pokedex Number'].nunique()
```

The entries for the following pokemon need to be corrected:
 - Pikachu
 - Darmanitan
 - Pumpkaboo
 - Gourgeist

```python
pokemon_df[pokemon_df['Pokedex Number']==25]
```

```python
pokemon_df[pokemon_df['Pokedex Number']==25].drop([pokemon_df.index[1141], pokemon_df.index[1142], pokemon_df.index[1139]]).replace(['Cosplay Pikachu'], 'Pikachu')
```

```python
pokemon_df = pokemon_df.drop([pokemon_df.index[1141], pokemon_df.index[1142], pokemon_df.index[1139]])
pik = pokemon_df.loc[pokemon_df['Pokemon'] == 'Cosplay Pikachu'].index[0]
pokemon_df.loc[[pik], ['Pokemon']]=['Pikachu']
```

```python
pokemon_df[pokemon_df['Pokedex Number']==25]
```

Pikachu entry corrected and removed extra entries

```python
pokemon_df[pokemon_df['Pokedex Number']==555]
```

```python
dar1 = pokemon_df.loc[pokemon_df['Pokemon'] == 'Darmanitan Galarian Standard Mode'].index[0]
dar2 = pokemon_df.loc[pokemon_df['Pokemon'] == 'Darmanitan Galarian Zen Mode'].index[0]
dar3 = pokemon_df.loc[pokemon_df['Pokemon'] == 'Darmanitan Zen Mode'].index[0]
pokemon_df.loc[[dar1], ['Total', 'HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']]=[480, 105, 140, 55, 30, 55, 95]
pokemon_df.loc[[dar2], ['Total', 'HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']]=[540, 105, 160, 55, 30, 55, 135]
pokemon_df.loc[[dar3], ['Total', 'HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']]=[540, 105, 30, 105, 140, 105, 55]
```

```python
pokemon_df[pokemon_df['Pokedex Number']==555]
```

Darmanitan entries have been corrected and now the missing sizes for Gourgeist and Pumpkaboo will be added and corrected

```python
gourgeist = pokemon_df.loc[pokemon_df['Pokemon'] == 'Gourgeist Super Size'].index[0]
pokemon_df.loc[[gourgeist], ['Pokemon']]=['Gourgeist Average Size']
missing_sizes = {
    'Pokedex Number': [711, 711, 711],
    'Pokemon': ['Gourgeist Small Size', 'Gourgeist Large Size', 'Gourgeist Super Size'],
    'Type 1': ['Ghost', 'Ghost', 'Ghost'],
    'Type 2': ['Grass', 'Grass', 'Grass'],
    'Total': [494, 494, 494],
    'HP': [55, 75, 85],
    'Attack': [85, 95, 100],
    'Defense': [122, 122, 122],
    'Sp. Atk': [58, 58, 58],
    'Sp. Def': [75, 75, 75],
    'Speed': [99, 69, 54],
    'Generation': [6, 6, 6],
    'Group': ['Standard', 'Standard', 'Standard']
}
missing_df = pd.DataFrame(missing_sizes)
pokemon_df = pd.concat([pokemon_df, missing_df], axis=0)
```

```python
pokemon_df[pokemon_df['Pokedex Number']==711]
```

Gourgeist now corrected

```python
pumpkaboo = pokemon_df.loc[pokemon_df['Pokemon'] == 'Pumpkaboo Super Size'].index[0]
pokemon_df.loc[[gourgeist], ['Pokemon']]=['Pumpkaboo Average Size']
missing_sizes = {
    'Pokedex Number': [710, 710, 710],
    'Pokemon': ['Pumpkaboo Small Size', 'Pumpkaboo Large Size', 'Pumpkaboo Super Size'],
    'Type 1': ['Ghost', 'Ghost', 'Ghost'],
    'Type 2': ['Grass', 'Grass', 'Grass'],
    'Total': [335, 335, 335],
    'HP': [44, 54, 59],
    'Attack': [66, 66, 66],
    'Defense': [70, 70, 70],
    'Sp. Atk': [44, 44, 44],
    'Sp. Def': [55, 55, 55],
    'Speed': [56, 46, 41],
    'Generation': [6, 6, 6],
    'Group': ['Standard', 'Standard', 'Standard']
}
missing_df = pd.DataFrame(missing_sizes)
pokemon_df = pd.concat([pokemon_df, missing_df], axis=0)
```

```python
pokemon_df[pokemon_df['Pokedex Number']==710]
```

Pumpkaboo now corrected

```python
pokemon_df = pokemon_df.sort_values(by=['Pokedex Number', 'Total'], ascending=True)
```
